```python
import pandas as pd
import numpy as np
```

#### read input file

```python
df = pd.read_csv("../../input/question-1/main.csv")
```

```python
df.head()
```

```python
df=df.drop(columns=['Total'])
```

```python
df.head()
```

```python
def sumUtil(df,start,end):
    df1=df.loc[start:end,"Population":].sum()
    df1["Year"]=df["Year"].iloc[start]
    return df1
```

```python
def process(df):
    start=0
    end=9
    n=df.shape[0]//10
    new_df=pd.DataFrame()
    for i in range(0,n):
        df1=sumUtil(df,start,end)
        new_df=new_df.append(df1,ignore_index=True)
        start=start+10
        end=end+9
        new_df= new_df.loc[:,:].astype('int64')
        
    df1=sumUtil(df,start,end)
    new_df=new_df.append(df1,ignore_index=True)
    new_df=new_df.set_index("Year")
    return new_df
    
```

```python
result = process(df)
```
